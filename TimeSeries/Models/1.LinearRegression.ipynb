# Régression Linéaire simple
## 1. Formulation mathématique
---
### 1.1 Enonciation du problème
Dans toute la suite, on notera que $\mathbb{K}$ désigne $\mathbb{R}$ ou $\mathbb{C}$. La régression linéaire simple se réalise sur un ensemble de données à deux variables tel que :

$$
Y = \beta_{0}X + \beta_{1}I_{n,1} + \epsilon \textcolor{white}{..}(1)
$$
ou :

1. $X = (x_{0}, ... ,x_{d}) \in \mathbb{K}^{d}$, variable explicative 
2. $Y = (y_{0}, ... ,y_{d})\in \mathbb{K}^{d}$, variable dépendante
3. $\epsilon = (\epsilon_{0}, ..., \epsilon_{d}) \in \mathbb{K}^{d}$ désigne un bruit
4. $I_{n,1} = (1, ..., 1)$ vecteur neutre
5. Les $(\beta_{i})_{i \in ⟦1; 2⟧}$ sont des coefficients à déterminer

$(P^{(1)})$ **L'objectif est de déterminer les coefficients $(\beta_{i})_{i \in ⟦1; 2⟧}$ qui minimisent l'erreur moyenne quadratique (MSE)** définie par :

$$
MSE(Y) = \frac{1}{n}\sum_{i = 1}^{n}(y_{i} - \hat{y_{i}})^{2}
$$

On cherche donc à estimer les $(\hat{y_{i}})_{i \in ⟦1; d⟧}$ tel que l'on retrouve l'équation $(1)$ grâce aux $(\beta_{i})_{i \in ⟦1; 2⟧}$. Il vient :

$$
MSE(Y) = MSE(\hat{\beta_{0}}, \hat{\beta_{1}})= \frac{1}{n}\sum_{i = 1}^{n}(y_{i} - \hat{\beta_{0}}x_{i} + \hat{\beta_{1}})^{2}
$$
Le problème $(P^{(1)})$ se ramène à un problème **d'optimisation convexe** :

$$
min_{(\beta_{i})_{i \in ⟦1; 2⟧}} \frac{1}{n}\sum_{i = 1}^{n}(y_{i} - \hat{\beta_{0}}x_{i} + \hat{\beta_{1}})^{2}
$$

---
### 1.2 Méthode  des moindres carrés
Pour résoudre $(P^{(1)})$, on utilise la méthode des **moindres carrés**, qui consiste à ce que les dérivées partielles 

$$
\frac{\partial MSE(\hat{\beta_{0}}, \hat{\beta_{1}})}{\partial \hat{\beta_{0}}} = 0
$$
$$
\frac{\partial MSE(\hat{\beta_{0}}, \hat{\beta_{1}})}{\partial \hat{\beta_{1}}} = 0
$$

Ce qui donne le système :

$$
\sum_{i = 0}^{n} x_{i}(y_{i} - \hat{\beta_{0}}x_{i} - \hat{\beta_{1}}) = 0
$$

$$
\sum_{i = 0}^{n}(y_{i} - \hat{\beta_{0}}x_{i} - \hat{\beta_{1}}) = 0
$$

En résolvant ce système on a les expressions des $(\hat{\beta_{i}})_{i \in ⟦1; 2⟧}$ :


$$
\hat{\beta_{0}} = \frac{n \sum_{i = 0}^{n} x_i y_i - \sum_{i = 0}^{n} x_i \sum_{i = 0}^{n} y_i}{n \sum_{i = 0}^{n} x_i^2 - (\sum_{i = 0}^{n} x_i)^2}
$$

$$
\hat{\beta_{1}} = \frac{1}{n}\sum_{i = 0}^{n}y_{i} - \frac{n \sum_{i = 0}^{n} x_i y_i - \sum_{i = 0}^{n} x_i \sum_{i = 0}^{n} y_i}{n \sum_{i = 0}^{n} x_i^2 - (\sum_{i = 0}^{n} x_i)^2}\frac{1}{n}\sum_{i = 0}^{n}x_{i}
$$

Remarque : ces deux relations montrent que le modèle est **consistant**, i.e. les coefficients $(\hat{\beta_{i}})_{i \in ⟦1; 2⟧}$ tendent vers $(\beta_{i})_{i \in ⟦1; 2⟧}$ quand $n$ tend vers $+\infty$.

#### 1.4.2 Coefficient de détermination $R^{2}$
Le coefficient de détermination est une mesure de la performance d'un modèle statistique, il justifie la relation entre la variable dépendante et les variables explicatives. Il est définit par le complément à 1 du rapport entre la some des carrés des résidus et la somme des carrés de l'écart entre les valeurs et la moyenne.

$$
R^2 = 1 - \frac{\sum_{i = 1}^{n} (y_i - \hat{y}_i)^2}{\sum_{i = 1}^{n} (y_i - \bar{y})^2}
$$

#### 1.4.3 Erreur quadratique moyenne $MSE$

L'erreur moyenne quadratique pour un signal $Y$ de taille $n$ est définie par :

$$
MSE  = \frac{1}{n}\sum_{i = 1}^{n}(y_{i} - \hat{y_{i}})^{2}
$$

1. Les $(y_{i})_{i \in ⟦1; n⟧}$ sont les valeurs réelles du, observables *a posteriori*
2. Les $(\hat{y_{i}})_{i \in ⟦1; n⟧}$ sont les valeurs estimées à l'aide du modèle (ici $\hat{y_{i}} = \beta_{0}x_{i} + \beta_{1}$)

Remarque : on prend parfois la racine carrée de la MSE, notée RMSE :

$$
RMSE  = \sqrt{\frac{1}{n}\sum_{i = 1}^{n}(y_{i} - \hat{y_{i}})^{2}}
$$

#### 1.4.4 Erreur d'estimation $\epsilon_{\hat{\beta}}$ sur les coefficients

A *posteriori*, l'erreur d'estimation sur les coefficients $(\beta_{i})_{i \in ⟦1; 2⟧}$ est représentée par le vecteur $\epsilon_{\beta} = \beta - \hat{\beta}$. L'erreur d'estimation est ensuite quantifiée par sa norme 2, i.e.

$$
|| \epsilon_{\beta}|| = || \beta - \hat{\beta} || = \sqrt{\sum_{i = 1}^{n}(\beta_{i} - \hat{\beta_{i}})}
$$
On peut néanmoins prendre d'autres mesures de l'erreur d'estimation, comme la norme $L^{p}$.

$$

# Régression linéaire multiple
---
## 1. Formulation mathématique
La régression linéaire multiple est une généralisation de la régression linéaire simple, ou une suite $(X_{n})_{n \in \mathbb{N}}$ de variables explicatives expliquent une variable dépendante $Y$, de telle sorte que $Y$ soit une combinaison linéaire des $(X_{n})_{n \in \mathbb{N}}$, i.e.

$$
Y = \beta_{0}I_{n,1} + \beta_{1}X_{1} + ... + \beta_{n}X_{n} + \epsilon  = \epsilon + \sum_{n = 0}^{N}\beta_{n}X_{n} \textcolor{white}{..}(2)
$$

ou :

1. $X_{i} = (x_{i,0}, ... ,x_{i,d}) \in \mathbb{K}^{d}$, variable explicative $i$
2. $Y = (y_{0}, ... ,y_{d})\in \mathbb{K}^{d}$, variable dépendante
3. $\epsilon = (\epsilon_{0}, ..., \epsilon_{d}) \in \mathbb{K}^{d}$ désigne un bruit
4. $I_{n,1} = (1, ..., 1)$ vecteur neutre
5. Les $(\beta_{i})_{i \in \mathbb{N}}$ sont des coefficients à déterminer

Soit :

$$
Y = X\beta + \epsilon =    
   \begin{bmatrix}
    x_{1,1} & \dots & x_{1,n} \\
    \vdots & x_{i,n} & \vdots \\
    x_{n,1} & \dots & x_{n,n} \\
   \end{bmatrix}
    \begin{bmatrix}
    \beta_{0} \\
    \vdots \\
    \beta_{n} \\
   \end{bmatrix} +
    \begin{bmatrix}
    \epsilon_{0} \\
    \vdots \\
    \epsilon_{n} \\
   \end{bmatrix}
$$

Comme pour la régression linéaire simple, $(P^{(2)})$ **L'objectif est de déterminer les coefficients $(\beta_{i})_{i \in \mathbb{N}}$ qui minimisent l'erreur moyenne quadratique (MSE)** définie par :

$$
\begin{align}
MSE(Y)  &= \frac{1}{n}\sum_{i = 1}^{N}(y_{i} - \hat{y_{i}})^{2} \\ 
        &= \frac{1}{n}\sum_{i = 1}^{N}(y_{i} - \sum_{n = 0}^{N}\hat{\beta_{n}}X_{n})^{2}
\end{align}
$$

Soit : 

$$
MSE((\beta_{i})_{i \in \mathbb{N}}) = ||Y - X\beta||^{2}
$$

Avec $X^{T}$ l'opérateur transpose dans l'espace des matrices à coefficients dans $\mathbb{K}$.
On cherche donc à avoir les $(\beta_{i})_{i \in \mathbb{N}}$ tel que :

$$
min_{\beta \in \mathbb{R}^{n}} ||Y - X\beta||^{2}
$$

Soit : 

$$
\mathbf{\beta} = (\mathbf{X}^T \mathbf{X})^{-1} \mathbf{X}^T \mathbf{Y}
$$

Les méthodes de résolution et d'évaluation des performances pour la régression simple sont étendues en dimension $n$.

#### 1.4.2 Erreur moyenne quadratique $ELMQ$ ou $MSE$

La MSE est un nombre entre -1 et 1 qui quantifie l'écart entre la valeur prédite $\hat{y_{i}}$ et sa vraie valeur $y_{i}$ mise au carré puis sommé et normalisée pour en faire sa moyenne.

$$

$$

On peut aussi définir sa racine carré pour obtenir un nombre entre 0 et 1 :

$$
RMSE = \sqrt{\frac{1}{n}\sum_{i = 1}^{N}(y_{i} - \hat{y_{i}})^{2}}
$$

\begin{bmatrix}
0.02 \\
0.01 \\
0.01
\end{bmatrix}
$$

L'erreur d'estimation est ensuite quantifiée par sa norme 2, i.e.

$$
|| \epsilon_{\beta} || = 0.024
$$
Comme pour l'erreur quadratique moyenne, plus $\epsilon_{\beta}$ est proche de 0 et plus le modèle est performant. Avec un $R^{2} = 0.8$, on s'attendait à avoir une erreur d'estimation faible.

Le modèle ressemble finalement à un plan, d'équation cartésienne $\hat{Y}$ :

```python
fig = px.scatter_3d(df, x='X1', y='X2', z='Y', color='Dataset')

# Génération de la surface de régression (plan) basé sur les coefficients du modèle
x1 = np.linspace(X[:, 0].min(), X[:, 0].max(), 20)
x2 = np.linspace(X[:, 1].min(), X[:, 1].max(), 20)
x1, x2 = np.meshgrid(x1, x2)

# Extraire le scalaire de l'array
beta0_scalar = beta0.item()  # ou beta0[0]

# Calcul du plan de régression
y_pred_grid = beta0_scalar + beta[0] * x1 + beta[1] * x2


# Ajouter le plan de régression
fig.add_trace(go.Surface(
    x=x1,
    y=x2,
    z=y_pred_grid,
    surfacecolor=np.zeros_like(y_pred_grid),  # surface "plate" pour le color mapping
    colorscale=[[0, 'orange'], [1, 'orange']],  # orange partout
    showscale=False,
    opacity=0.6
))

# Ajouter des axes 3D
fig.update_layout(
    scene=dict(
        xaxis_title='X1',
        yaxis_title='X2',
        zaxis_title='Y'
    ),
    title=f'Régression linéaire multiple : Y = {beta[0]:.2f}x + {beta[1]:.2f}y + {beta0[0]:.2f}'
)

fig.show() # Affichage du graphique
```

# Régressions régularisées

## Régression Ridge
---
### 1. Formulation mathématique
On reprend la même structure que la régression linéaire multiple. On prend une suite $(X_{n})_{n \in \mathbb{N}}$ de variables explicatives expliquent une variable dépendante $Y$, de telle sorte que $Y$ soit une combinaison linéaire des $(X_{n})_{n \in \mathbb{N}}$, i.e.

$$
Y = \beta_{0}I_{n,1} + \beta_{1}X_{1} + ... + \beta_{n}X_{n} + \epsilon  = \epsilon + \sum_{n = 0}^{N}\beta_{n}X_{n} \textcolor{white}{..}(2)
$$

ou :

1. $X_{i} = (x_{i,0}, ... ,x_{i,d}) \in \mathbb{K}^{d}$, variable explicative $i$
2. $Y = (y_{0}, ... ,y_{d})\in \mathbb{K}^{d}$, variable dépendante
3. $\epsilon = (\epsilon_{0}, ..., \epsilon_{d}) \in \mathbb{K}^{d}$ désigne un bruit
4. $I_{n,1} = (1, ..., 1)$ vecteur neutre
5. Les $(\beta_{i})_{i \in \mathbb{N}}$ sont des coefficients à déterminer

Soit :

$$
Y = X\beta + \epsilon =    
   \begin{bmatrix}
    x_{1,1} & \dots & x_{1,n} \\
    \vdots & x_{i,n} & \vdots \\
    x_{n,1} & \dots & x_{n,n} \\
   \end{bmatrix}
    \begin{bmatrix}
    \beta_{0} \\
    \vdots \\
    \beta_{n} \\
   \end{bmatrix} +
    \begin{bmatrix}
    \epsilon_{0} \\
    \vdots \\
    \epsilon_{n} \\
   \end{bmatrix}
$$

Comme pour la régression linéaire multiple, $(P^{(2)})$ **L'objectif est de déterminer les coefficients $(\beta_{i})_{i \in \mathbb{N}}$ qui minimisent l'erreur moyenne quadratique (MSE), cette fois-ci avec une pénalité** définie par :

$$
MSE((\beta_{i})_{i \in \mathbb{N}}) = ||Y - X\beta||^{2} + \lambda||\beta||^{2}
$$

Avec :
1. $X^{T}$ l'opérateur transpose dans l'espace des matrices à coefficients dans $\mathbb{K}$
2. $\lambda \in \mathbb{R}^{+}$ un paramètre de régularisation (**Hyperparamètre**).

On cherche donc à avoir les $(\beta_{i})_{i \in \mathbb{N}}$ tel que :

$$
min_{\beta \in \mathbb{R}^{n}} (||Y - X\beta||^{2} + \lambda||\beta||^{2})
$$

Soit : 

$$
\mathbf{\beta} = (\mathbf{X}^T \mathbf{X} + \lambda \mathbf{I_{n}})^{-1} \mathbf{X}^T \mathbf{Y}
$$

1. Si on a $\lambda = 0$, on retrouve la régression linéaire classique. Pour un $\lambda$ grand, les variables faiblement corrélées avec $Y$ vont avoir un effet négligeable sur la régression.
2. Les méthodes de résolution et d'évaluation des performances pour la régression ridge sont les mêmes que pour la régression multiple
