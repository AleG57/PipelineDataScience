# Régression Linéaire simple
---

## 1. Formulation mathématique
### 1.1 Enonciation du problème
La régression linéaire simple se réalise sur un ensemble de données à deux variables tel que :

$$
Y = \beta_{0}X + \beta_{1}I_{n,1} + \epsilon \textcolor{white}{..}(1)
$$
ou :

1. $X = (x_{0}, ... ,x_{d}) \in \mathbb{K}^{d}$, variable explicative 
2. $Y = (y_{0}, ... ,y_{d})\in \mathbb{K}^{d}$, variable dépendante
3. $\epsilon = (\epsilon_{0}, ..., \epsilon_{d}) \in \mathbb{R}^{d}$ désigne un bruit
4. $I_{n,1} = (1, ..., 1)$ vecteur neutre
5. Les $(\beta_{i})_{i \in ⟦1; 2⟧}$ sont des coefficients à déterminer

$(P^{(1)})$ **L'objectif est de déterminer les coefficients $(\beta_{i})_{i \in ⟦1; 2⟧}$ qui minimisent l'erreur moyenne quadratique (MSE)** définie par :

$$
MSE(Y) = \frac{1}{n}\sum_{i = 1}^{n}(y_{i} - \hat{y_{i}})^{2}
$$

On cherche donc à estimer les $(\hat{y_{i}})_{i \in ⟦1; d⟧}$ tel que l'on retrouve l'équation $(1)$ grâce aux $(\beta_{i})_{i \in ⟦1; 2⟧}$. Il vient :

$$
MSE(Y) = MSE(\hat{\beta_{0}}, \hat{\beta_{1}})= \frac{1}{n}\sum_{i = 1}^{n}(y_{i} - \hat{\beta_{0}}x_{i} + \hat{\beta_{1}})^{2}
$$
Le problème $(P^{(1)})$ se ramène à un problème **d'optimisation convexe** :

$$
min_{(\beta_{i})_{i \in ⟦1; 2⟧}} \frac{1}{n}\sum_{i = 1}^{n}(y_{i} - \hat{\beta_{0}}x_{i} + \hat{\beta_{1}})^{2}
$$

### 1.2 Méthode  des moindres carrés (OLS) 

Pour résoudre $(P^{(1)})$, on utilise la méthode des **moindres carrés**, qui consiste à ce que les dérivées partielles 

$$
\frac{\partial MSE(\hat{\beta_{0}}, \hat{\beta_{1}})}{\partial \hat{\beta_{0}}} = 0
$$
$$
\frac{\partial MSE(\hat{\beta_{0}}, \hat{\beta_{1}})}{\partial \hat{\beta_{1}}} = 0
$$

Ce qui donne le système :

$$
\sum_{i = 0}^{n} x_{i}(y_{i} - \hat{\beta_{0}}x_{i} - \hat{\beta_{1}}) = 0
$$

$$
\sum_{i = 0}^{n}(y_{i} - \hat{\beta_{0}}x_{i} - \hat{\beta_{1}}) = 0
$$

En résolvant ce système on a les expressions des $(\hat{\beta_{i}})_{i \in ⟦1; 2⟧}$ :


$$
\hat{\beta_{0}} = \frac{n \sum_{i = 0}^{n} x_i y_i - \sum_{i = 0}^{n} x_i \sum_{i = 0}^{n} y_i}{n \sum_{i = 0}^{n} x_i^2 - (\sum_{i = 0}^{n} x_i)^2}
$$

$$
\hat{\beta_{1}} = \frac{1}{n}\sum_{i = 0}^{n}y_{i} - \frac{n \sum_{i = 0}^{n} x_i y_i - \sum_{i = 0}^{n} x_i \sum_{i = 0}^{n} y_i}{n \sum_{i = 0}^{n} x_i^2 - (\sum_{i = 0}^{n} x_i)^2}\frac{1}{n}\sum_{i = 0}^{n}x_{i}
$$

Remarque : ces deux relations montrent que le modèle est **consistant**, i.e. les coefficients $(\hat{\beta_{i}})_{i \in ⟦1; 2⟧}$ tendent vers $(\beta_{i})_{i \in ⟦1; 2⟧}$ quand $n$ tend vers $+\infty$.

## 2. Des exemples en Python
---
#### 2.1 Import des librairies

```python
# Modules math
import numpy as np
import random as rd
import pandas as pd
import matplotlib.pyplot as plt

# Modules pour la régression linéaire
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

# Sauvegarde du modèle
from joblib import dump, load

import seaborn as sns # Style
```

#### 2.2 Initialisation / import des données
Ici, on prend pour l'exemple d'un dataset de telle sorte que :

$$
Y = 2X + 3 + \epsilon
$$

Ou $\epsilon = (\epsilon_{0}, ..., \epsilon_{d})$ désigne un bruit blanc centré ergodique de variance 1, c'est à dire que les $(\epsilon_{i})_{i \in \mathbb{N}}$ sont des variables aléatoires qui suivent une loi normale de paramètre $(0,1)$. Dit autrement, on force une relation linéaire entre les deux varibles en rajoutant du bruit pour simuler un vrai dataset. 

```python
# Init
sns.set_style("darkgrid") # Style du graphe
N = 1000

# Init X et Y (Y = 2*X + 3 + bruit)
X = np.array([
   np.random.normal(0,1) for _ in range(N)
]).reshape(-1,1)
Y = 2*X + 3 + np.array([
   np.random.normal(0,1) for _ in range(N)
]).reshape(-1,1)

# Visualisation du dataset
df = pd.DataFrame({
    "y" : Y.reshape(-1),
    "X" : X.reshape(-1)
})
df.head()
```

#### 2.3 Séparation des données de test et d'entraînement

On sépare les données de test et d'entraînement pour connaître la fiabilité du modèle. L'idée est de séparer le datset initial en deux parties (datasets) distinctes grâce à la méthode train\_test\_split. Les arguments de la méthode sont :

1. test\_size : proportion (%) des données de test. Par exemple, si test\_size = 0.2, la méthode va prendre 20% des données comme étant du test et les 80% restants seront alloués à l'entraînement du modèle.
2. random\_state : c'est la seed de notre modèle. Cet argument permet de le sauvegarder et de le rendre unique pour pouvoir le réutiliser plus loin dans le code.

L'intérêt de cette méthode est de pouvoir observer le comportement du modèle sur un dataset qu'il ne connaît pas afin de calculer son ELMQ.

```python
# Separating the data into data to entertain and row data in order to comparate trained data and row data (performance test)
# test_size : proportion de données test (20%)
# random_state : random seed
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.2, random_state=42)

# Visualisation des données
plt.scatter(X_train,Y_train, label="Données d'entraînement", color="Orange")
plt.scatter(X_test,Y_test, label = "Données de test", color="Blue")
plt.legend()
plt.title("Données brutes")
plt.show()
```

Comme on peut le voir sur ce graphique, en orange les données de "train" et en bleu les données de "test", inconnues du modèle.

#### 2.4 Entraînement du modèle

```python
lr = LinearRegression() # Initialisation du modèle
lr.fit(X_train,Y_train) # Entraînement du modèle
```

#### 2.5 Prédictions

```python
y_pred = lr.predict(X_test) # Prédicitons sur les données de test
```

#### 2.6 Evaluation des performances du modèle

On observe le comportement du modèle sur les données de test uniquement, c'est à dire la partie du dataset inconnu au modèle, on appelle cela **l'inférence**. Ensuite on détermine son coefficient de détermination $R^{2}$, soit l'indicateur de performance du modèle. Plus le $R^{2}$ est proche de 1, plus le modèle est fiable. Remarque : si on prend le coefficient $R$ de base, le signe de $R$ donne une information sur l'orientation de la pente. Si $R < 0$, alors la relation entre les deux variables est décroissante. En revanche si $R > 0 $, alors la relation est croissante.

```python
y_pred_train = lr.predict(X_train) # Prédicitons sur les données de train (pour l'exemple)

R2 = lr.score(X_train,Y_train) # R² pour l'évaluation des performances 
MSE = mean_squared_error(Y_train, y_pred_train) # Mean Squared Error, autre évalution de performance

# Affichage de la régression sur les données inconnues au modèle
plt.scatter(X_train,Y_train, label = "Données de train", color = "orange")
plt.plot(X_train, y_pred_train, color="black", label=f"Régression. R^2 = {R2:.1f}, MSE = {MSE:.1f}")
plt.legend()
plt.title("Régression linéaire sur les données train")
plt.show()
```

```python
R2 = lr.score(X_test,Y_test) # R² pour l'évaluation des performances 
MSE = mean_squared_error(Y_test, y_pred) # Mean Squared Error, autre évalution de performance

# Affichage de la régression sur les données inconnues au modèle
plt.scatter(X_test,Y_test, label = "Données de test", color = "Blue")
plt.plot(X_test, y_pred, color="black", label=f"Régression. R^2 = {R2:.1f}, MSE = {MSE:.1f}")
plt.legend()
plt.title("Régression linéaire sur les données test")
plt.show()
```

On remarque que le $R^{2}$ est proche de 1 et la MSE proche de 0 pour les deux cas. On pouvait s'y attendre pour les données de train, ce qui n'est pas forcément le cas pour les données de test car elles sont inconnues au modèle : elles peuvent êtres radicalement différentes des données train si elles ne sont pas choisies correcement. Le modèle est donc fiable.

#### 2.7 Equation de la droite

Pour rappel, l'objectif était d'estimer les coefficients $\hat{\beta_{0}}$ et $\hat{\beta_{1}}$ qui satisfont l'équation cartésienne de la courbe représentative en noir du graphique ci-dessus.  

```python
beta0 = lr.coef_[0] # Coefficient directeur
beta1 = lr.intercept_ # Ordonnée à l'origine
```

```python
beta0
```

```python
beta1
```

L'équation cartésienne de la courbe noire est donc 

$$
\hat{Y} = 2.04X + 2.99
$$

Ce qui correspond assez bien à l'équation que l'on doit retrouver :

$$
Y = 2X + 3
$$

A *posteriori*, l'erreur d'estimation sur les coefficients $(\beta_{i})_{i \in ⟦1; 2⟧}$ est représentée par le vecteur $\epsilon_{\beta} = \beta - \hat{\beta}$. Dans cet exemple, on peut le calculer aisément.

$$ \epsilon_{\beta} = 
\begin{bmatrix}
2 \\
3
\end{bmatrix} - 
\begin{bmatrix}
2.04 \\
2.99
\end{bmatrix} =
\begin{bmatrix}
0.04 \\
0.01
\end{bmatrix}
$$

L'erreur d'estimation est ensuite quantifiée par sa norme 2, i.e.

$$
|| \epsilon_{\beta} || = 0.041
$$
Comme pour l'erreur quadratique moyenne, plus $\epsilon_{\beta}$ est proche de 0 et plus le modèle est performant. Avec un $R^{2} = 0.8$, on s'attendait à avoir une erreur d'estimation faible.

Le modèle ressemble finalement à une droite d'équation cartésienne $\hat{Y} = 2.05X + 2.99$ :

```python
plt.scatter(X_train,Y_train, label="Données d'entraînement", color="Orange")
plt.scatter(X_test,Y_test, label = "Données de test", color="Blue")
plt.plot(X_test, y_pred, color="black", label=f"Régression. R^2 = {R2:.1f}")

plt.legend()
plt.title(f"Equation de la régression linéaire : Y = {beta0[0]:.2f}X + {beta1[0]:.2f}")
plt.show()
```

#### 2.8 Déploiement du modèle 

Une fois avoir confirmé que le modèle est fiable, il reste à le déployer. Pour cela, on le sauvegarde pour pouvoir le redéployer dans un autre cas.

```python
dump(lr, 'RegressionLineaireSimple.joblib') # Enregistrement du modèle
```

Une fois le modèle sauvegardé, on peut le charger pour plus tard.

```python
model = load('RegressionLineaireSimple.joblib') # Recharge du modèle
y_pred = model.predict(X_test) # Faire une prédiction
```

# Régression linéaire multiple
---

## 1.1 Rappel de la définition d'un dataset

Pour toute la suite, on considère un *dataset* $\mathcal{D}$ de $n$ échantillons, $J$ variables, i.e. 

$$
\mathcal{D} \stackrel{\text{def}}:=
\{ (x_i, y_i) \quad | \quad x_i, y_i \in \mathcal{X} \times \mathcal{Y}, \quad \forall i \in {\mathbb{N}} \}
$$

Avec : 

* $x_i \in \mathbb{R}^n, X_i = (x_{i,1}, \dots, x_{i,n})$ : $i$-ème échantillon du dataset. Les réels $x_{i,j}$ représentent une valeur de chaque *features* $j$. 
* $y_i \in \mathbb{R}$ : $i$-ème réalisation de la variable cible

Ainsi, $\mathcal{D} = (X, y)$ avec la *design matrix* $\mathbb{X} = (X_i)_{i \in \{1;n\}}$ et $y = (y_i)_{i \in \{1;n\}}$

$$
\mathbb{X} = 
\begin{bmatrix}
x_{1,1} & \dots & x_{1,J} \\
\vdots & x_{i,j} & \vdots \\
x_{n,1} & \dots & x_{n,J} \\
\end{bmatrix} \quad y = 
\begin{bmatrix}
y_{0} \\
\vdots \\
y_{n} \\
\end{bmatrix}
$$

## 1.2 Formulation mathématique

La régression linéaire multiple est une généralisation de la régression linéaire simple, ou une suite $(X_{n})_{n \in \mathbb{N}}$ de $N$ variables expliquent une variable dépendante $Y$ de telle sorte que $Y$ soit une combinaison linéaire des $(X_{n})_{n \in \mathbb{N}}$, i.e.

$$
Y = \beta_{0}I_{n,1} + \beta_{1}X_{1} + ... + \beta_{n}X_{n} + \epsilon  = \sum_{n = 0}^{N}\beta_{n}X_{n} + \epsilon \quad (2)
$$

Avec :

* $X_{i} = (x_{i,0}, ... ,x_{i,d}) \in \mathbb{R}^{d}$, variable explicative $i$
* $Y = (y_{0}, ... ,y_{d})\in \mathbb{R}^{d}$, variable dépendante
* $\epsilon = (\epsilon_{0}, ..., \epsilon_{d}) \in \mathbb{R}^{d}$ désigne un bruit. On fait l'hypothèse que le bruit est centré suivant les données d'entrées, i.e.  $\mathbb{E}(\epsilon | X) = \mathbb{E}(\epsilon | X_1, \dots, X_N) = 0$. On lui suppose aussi l'homoscédasticité (variance constante), i.e. $\mathbb{V}(\epsilon | X) = \mathbb{V}(\epsilon | X_1, \dots, X_N) = \sigma_{\epsilon}^2 \in \mathbb{R}$. Sinon, le bruit est considéré hétéroscédastique. 
* $I_{n,1} = (1, ..., 1)$ vecteur neutre
* Les $(\beta_{i})_{i \in \mathbb{N}}$ sont des coefficients à déterminer avec $\beta = (\beta_0, \dots, \beta_d) \in \mathbb{R}^d$

A partir des *features*, $X_i$, on construit la *design matrix* notée $\mathbb{X}$, soit :

$$
Y = \mathbb{X}\beta + \epsilon =    
   \begin{bmatrix}
    x_{1,1} & \dots & x_{1,n} \\
    \vdots & x_{i,n} & \vdots \\
    x_{n,1} & \dots & x_{n,n} \\
   \end{bmatrix}
    \begin{bmatrix}
    \beta_{0} \\
    \vdots \\
    \beta_{n} \\
   \end{bmatrix} +
    \begin{bmatrix}
    \epsilon_{0} \\
    \vdots \\
    \epsilon_{n} \\
   \end{bmatrix}
$$

Comme pour la régression linéaire simple, $(P^{(2)})$, l'objectif est de déterminer les coefficients $(\beta_{i})_{i \in \mathbb{N}}$ qui minimisent l'erreur moyenne quadratique (MSE) :
$$
\boxed{\min_{\beta \in \mathbb{R}^{n}} ||Y - \mathbb{X}\beta||^{2}}
$$

Ce problème d'optimisation est connu sous le nom d'OLS (Ordinary Least Squares). La solution de ce problème est sous la forme : 

$$
\beta = (\mathbb{X}^T \mathbb{X})^{-1} \mathbb{X}^T Y \quad (i)
$$

Intuitivement, l'objectif est de minimiser la distance entre l'estimateur de $Y$, $\hat{Y}$ et $Y$. Graphiquement, ce problème peut se tracer de la forme suivante : 

<div align="center">
  <img src="src/pics/LR/OLSinterpretation.png" alt="a" width="500" height="350">
</div>

Dans cette interpretation, on note $\hat{\beta} = \hat{\theta}^T$ et $X = \mathbb{X}, N = m$

### 1.2 Méthode de descente du gradient

Bien que la solution analytique soit donnée par l'équation précédente $(i)$, cela suppose d'inverser une matrice de la taille du *dataset*. Or, même en utilisant les algorithme les plus performants d'inversion de matrices, la complexité de cet algorithme est en $\mathcal{O}(n^3)$. On propose donc une solution approchée qui permet de diminuer drastiquement la temps de calcul de cette solution avec de très bonnes approximations.

## 2. Un exemple en Python
---
### 2.1 Import des librairies nécéssaires

```python
# Modules math
import numpy as np
import random as rd
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objects as go

# Modules pour la régression linéaire
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from joblib import dump, load

import seaborn as sns # Style

import plotly.express as px # Intégration des graphiques 3D dans jupyter
```

### 2.2 Initialisation / import des données 
Ici, pour des problèmes de compréhension et de visualisation, on prendra l'exemple d'un dataset de deux variables explicatives, $X1$ et $X2$ et d'une variable dépendante $Y$ tel que :

$$
Y = 2X_{1} + 3X_{2} - 1
$$

Ou $\epsilon = (\epsilon_{0}, ..., \epsilon_{d})$ désigne un bruit blanc centré ergodique de variance 1, c'est à dire que les $(\epsilon_{i})_{i \in \mathbb{N}}$ sont des variables aléatoires qui suivent une loi normale de paramètre $(0,1)$. Autremeent dit, $Y$ est une combinaison linéaire de $X_{1}$ et $X_{2}$ plus un bruit, toujours pour simuler un vrai dataset.

```python
# Init X et Y (Y = 2*X + 3 + bruit)
X1 = np.array([
   np.random.normal(0,1) for _ in range(N)
]).reshape(-1,1)
X2 = np.array([
   np.random.normal(0,1) for _ in range(N)
]).reshape(-1,1)

Y = 2*X1 + 3*X2 - 1 + np.array([
   np.random.normal(0,0.5) for _ in range(N)
]).reshape(-1,1)

df = pd.DataFrame({
    "Y" : Y.reshape(-1),
    "X1" : X1.reshape(-1),
    "X2" : X2.reshape(-1)
})
df.head()
```

### 2.3 Séparation des données de test et d'entraînement

La seule différence avec la régression linéaire est que les variable explicatives (features) doivent êtres stackées avant de séparer les deux vecteurs en deux datasets de test et d'entraînement distincts.

```python
X = np.column_stack((X1, X2)) # Stack des features
```

On retrouve alors la même méthode utilisée que pour la régression linéaire simple :

On sépare les données de test et d'entraînement pour connaître la fiabilité du modèle. L'idée est de séparer le datset initial en deux parties (datasets) distinctes grâce à la méthode train\_test\_split. Les arguments de la méthode sont :

1. test\_size : proportion (%) des données de test. Par exemple, si test\_size = 0.3, la méthode va prendre 30% des données comme étant du test et les 80% restants seront alloués à l'entraînement du modèle.
2. random\_state : c'est la seed de notre modèle. Cet argument permet de le sauvegarder et de le rendre unique pour pouvoir le réutiliser plus loin dans le code.

L'intérêt de cette méthode est de pouvoir observer le comportement du modèle sur un dataset qu'il ne connaît pas afin de calculer son ELMQ.

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
# Visualisation des données
fig = px.scatter_3d(df, x='X1', y='X2', z='Y', color='Y', color_continuous_scale='viridis')
fig.show()

# Séparation des datasets train et test pour la visualisation
df_train = pd.DataFrame({
    'X1': X_train[:, 0].ravel(),
    'X2': X_train[:, 1].ravel(),
    'Y': Y_train.ravel(),
    'Dataset': 'Train'
})

df_test = pd.DataFrame({
    'X1': X_test[:, 0].ravel(),
    'X2': X_test[:, 1].ravel(),
    'Y': Y_test.ravel(),
    'Dataset': 'Test'
})

# Fusion des datasets 
df = pd.concat([df_train, df_test])

# Graphique 3D 
fig = px.scatter_3d(df, x='X1', y='X2', z='Y', color='Dataset', 
                    color_discrete_map={'Train': 'blue', 'Test': 'red'})

fig.show()
```

Comme on peut le voir sur ce graphique, en orange les données de "train" et en bleu les données de "test", inconnues du modèle.

### 2.4 Entraînement du modèle

```python
lr = LinearRegression() # Initialisation du modèle
lr.fit(X_train,Y_train) # Entraînement du modèle
```

### 2.5 Prédictions

```python
y_pred = lr.predict(X_test) # Prédicitons sur les données de test
```

### 2.6 Evaluation des performances du modèle

Une fois les variables explicatives stackées, le modèle se comporte comme une régression linéaire simple. On s'en remet alors 

On observe le comportement du modèle sur les données de test uniquement, puis on détermine son coefficient de détermination $R^{2}$, soit l'indicateur de performance du modèle. Plus le $R^{2}$ est proche de 1, plus le modèle est fiable. Remarque : si on prend le coefficient $R$ de base, le signe de $R$ donne une information sur l'orientation de la pente. Si $R < 0$, alors la relation entre les deux variables est décroissante. En revanche si $R > 0 $, alors la relation est croissante.

```python
R2 = lr.score(X_test,Y_test) # R²
MSE = mean_squared_error(Y_test, y_pred) # MSE 
```

```python
f"{MSE:.2f}"
```

```python
f"{R2:.2f}"
```

La $MSE$ est proche de 0 et le $R^{2}$ est très proche de 1, ce qui signifie que le modèle est très performant.

### 2.7 Equation de la droite

```python
beta = lr.coef_[0] # Coefficient directeur
beta0 = lr.intercept_ # Ordonnée à l'origine
```

```python
beta0
```

```python
beta
```

L'équation cartésienne de la courbe noire est donc 

$$
\hat{Y} = 1.98X1 + 3.01X2 - 0.99
$$

Ce qui correspond assez bien à l'équation que l'on doit retrouver :

$$
\hat{Y} = 2X_{1} + 3X_{2} - 1
$$

A *posteriori*, l'erreur d'estimation sur les coefficients $(\beta_{i})_{i \in ⟦1; 2⟧}$ est représentée par le vecteur $\epsilon_{\beta} = \beta - \hat{\beta}$. Dans cet exemple, on peut le calculer aisément.

$$ \epsilon_{\beta} = 
\begin{bmatrix}
2 \\
3 \\
-1
\end{bmatrix} - 
\begin{bmatrix}
1.98 \\
3.01 \\
-0.99
\end{bmatrix} =
\begin{bmatrix}
0.02 \\
0.01 \\
0.01
\end{bmatrix}
$$

L'erreur d'estimation est ensuite quantifiée par sa norme 2, i.e.

$$
|| \epsilon_{\beta} || = 0.024
$$
Comme pour l'erreur quadratique moyenne, plus $\epsilon_{\beta}$ est proche de 0 et plus le modèle est performant. Avec un $R^{2} = 0.8$, on s'attendait à avoir une erreur d'estimation faible.

Le modèle ressemble finalement à un plan, d'équation cartésienne $\hat{Y}$ :

```python
fig = px.scatter_3d(df, x='X1', y='X2', z='Y', color='Dataset')

# Génération de la surface de régression (plan) basé sur les coefficients du modèle
x1 = np.linspace(X[:, 0].min(), X[:, 0].max(), 20)
x2 = np.linspace(X[:, 1].min(), X[:, 1].max(), 20)
x1, x2 = np.meshgrid(x1, x2)

# Extraire le scalaire de l'array
beta0_scalar = beta0.item()  # ou beta0[0]

# Calcul du plan de régression
y_pred_grid = beta0_scalar + beta[0] * x1 + beta[1] * x2


# Ajouter le plan de régression
fig.add_trace(go.Surface(
    x=x1,
    y=x2,
    z=y_pred_grid,
    surfacecolor=np.zeros_like(y_pred_grid),  # surface "plate" pour le color mapping
    colorscale=[[0, 'orange'], [1, 'orange']],  # orange partout
    showscale=False,
    opacity=0.6
))

# Ajouter des axes 3D
fig.update_layout(
    scene=dict(
        xaxis_title='X1',
        yaxis_title='X2',
        zaxis_title='Y'
    ),
    title=f'Régression linéaire multiple : Y = {beta[0]:.2f}x + {beta[1]:.2f}y + {beta0[0]:.2f}'
)

fig.show() # Affichage du graphique
```

## Régression Ridge

### 1. Formulation mathématique

On reprend la même structure que la régression linéaire multiple. On prend une suite $(X_{n})_{n \in \mathbb{N}}$ de variables explicatives expliquent une variable dépendante $Y$, de telle sorte que $Y$ soit une combinaison linéaire des $(X_{n})_{n \in \mathbb{N}}$, i.e.

$$
Y = \beta_{0}I_{n,1} + \beta_{1}X_{1} + ... + \beta_{n}X_{n} + \epsilon  = \sum_{n = 0}^{N}\beta_{n}X_{n} + \epsilon \quad (2)
$$

Avec :

* $X_{i} = (x_{i,0}, ... ,x_{i,d}) \in \mathbb{R}^{d}$, variable explicative $i$
* $Y = (y_{0}, ... ,y_{d})\in \mathbb{R}^{d}$, variable dépendante
* $\epsilon = (\epsilon_{0}, ..., \epsilon_{d}) \in \mathbb{R}^{d}$ désigne un bruit. On fait l'hypothèse que le bruit est centré suivant les données d'entrées, i.e.  $\mathbb{E}(\epsilon | X) = \mathbb{E}(\epsilon | X_1, \dots, X_N) = 0$. On lui suppose aussi l'homoscédasticité (variance constante), i.e. $\mathbb{V}(\epsilon | X) = \mathbb{V}(\epsilon | X_1, \dots, X_N) = \sigma_{\epsilon}^2 \in \mathbb{R}$. Sinon, le bruit est considéré hétéroscédastique. 
* $I_{n,1} = (1, ..., 1)$ vecteur neutre
* Les $(\beta_{i})_{i \in \mathbb{N}}$ sont des coefficients à déterminer avec $\beta = (\beta_0, \dots, \beta_d) \in \mathbb{R}^d$

A partir des *features*, $X_i$, on construit la *design matrix* notée $\mathbb{X}$, soit :

$$
Y = \mathbb{X}\beta + \epsilon =    
   \begin{bmatrix}
    x_{1,1} & \dots & x_{1,n} \\
    \vdots & x_{i,n} & \vdots \\
    x_{n,1} & \dots & x_{n,n} \\
   \end{bmatrix}
    \begin{bmatrix}
    \beta_{0} \\
    \vdots \\
    \beta_{n} \\
   \end{bmatrix} +
    \begin{bmatrix}
    \epsilon_{0} \\
    \vdots \\
    \epsilon_{n} \\
   \end{bmatrix}
$$

Comme pour la régression linéaire multiple, $(P^{(2)})$ **L'objectif est de déterminer les coefficients $(\beta_{i})_{i \in \mathbb{N}}$ qui minimisent l'erreur moyenne quadratique (MSE) avec une pénalité $L_2$**, ce qui correspond au problème d'optimisation suivant :

$$
\min_{\beta}||Y - X\beta||^{2} + \lambda||\beta||^{2}
$$

Avec :
1. $X^{T}$ l'opérateur transpose dans l'espace des matrices à coefficients dans $\mathbb{K}$
2. $\lambda \in \mathbb{R}^{+}$ un paramètre de régularisation (**Hyperparamètre**).

On cherche donc à avoir les $(\beta_{i})_{i \in \mathbb{N}}$ tel que :

$$
\min_{\beta \in \mathbb{R}^{n}} (||Y - X\beta||^{2} + \lambda||\beta||^{2})
$$

Soit : 

$$
\mathbf{\beta} = (\mathbf{X}^T \mathbf{X} + \lambda \mathbf{I_{n}})^{-1} \mathbf{X}^T \mathbf{Y}
$$

1. Si on a $\lambda = 0$, on retrouve la régression linéaire classique. Pour un $\lambda$ grand, les variables faiblement corrélées avec $Y$ vont avoir un effet négligeable sur la régression.
2. Les méthodes de résolution et d'évaluation des performances pour la régression ridge sont les mêmes que pour la régression multiple

# Régression Linéaire Généralisée (GLM)
---

La régression linéaire généralisée permet d'appliquer une régression linéaire sur les paramètres $\beta$ tout en autorisant la relation entre les *features* $X_i$ et $Y$ comme étant linéaire ou non. On introduit donc une fonction $\Phi$ tel que :

$$
Y = f(\Phi(\mathbb{X})\beta + \epsilon) 
$$

Avec :

* $X_{i} = (x_{i,0}, ... ,x_{i,d}) \in \mathbb{R}^{d}$, variable explicative $i$
* $Y = (y_{0}, ... ,y_{d})\in \mathbb{R}^{d}$, variable dépendante
* $\epsilon = (\epsilon_{0}, ..., \epsilon_{d}) \in \mathbb{R}^{d}$ désigne un bruit. On fait l'hypothèse que le bruit est centré suivant les données d'entrées, i.e.  $\mathbb{E}(\epsilon | X) = \mathbb{E}(\epsilon | X_1, \dots, X_N) = 0$. On lui suppose aussi l'homoscédasticité (variance constante), i.e. $\mathbb{V}(\epsilon | X) = \mathbb{V}(\epsilon | X_1, \dots, X_N) = \sigma_{\epsilon}^2 \in \mathbb{R}$. Sinon, le bruit est considéré hétéroscédastique. 
* Les $(\beta_{i})_{i \in \mathbb{N}}$ sont des coefficients à déterminer avec $\beta = (\beta_0, \dots, \beta_d) \in \mathbb{R}^d$
* $\Phi$ une fonction.

On peut reconstruire la relation non-linéaire entre les $X_i$ et $Y$ par le changement de variables $X_i' = \Phi_i(X_i), \forall i \in \{1;N\}$
