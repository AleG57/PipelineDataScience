## 1. Formulation mathématique

Une random forest est un regrouppement de plusieurs arbres de décision basés sur la sélection aléatoire de variables explicatives (boostrap aggregating).

1. **Construction d'une forêt aléatoire**

On génère plusieurs jeux de données d'entraînement en effectuant un tirage aléatoire avec remise dans le dataset initial noté $\mathbb{D}_{i}$. On note $\mathcal{D}$ l'ensemble des jeux de données issu du boostrap :

$$
\mathcal{D} = \{ (D_{j})_{j \in \mathbb{N}} \} = \{ (Y, (X_{i})_{i \in \mathbb{N}}\}
$$

Pour chaque $(D_{i})_{i \in \mathbb{N}} \in \mathcal{D}$, on construit un arbre de décision $T_{i}$.

2. **Sélection des variables**

Pour chaque arbres, on re-sélectionne uniquement un sous ensemble des variables tirées aléatoirement à chaque noeud des arbres, basé sur une sélection avec des critères de séparation comme leurs entropie ou le critère de Gini.

3. **Prise de décision**

Chaque arbre $T_{i}$ prédit une classe $\hat{y}_{i} \in \mathbb{R}$. La méthode de sélection de la classe dépend du type de problème que l'on veut résoudre : une classification ou une régression. Si c'est une classification, on cherche à avoir la classe majoritaire tandis qu'ne problème de régression, on cherche à avoir une moyenne.

### 1.1 Random forest en classification

$$
\hat{y} = mode \{\hat{y_{1}}, ..., \hat{y_{n}} \}
$$

ou $mode$ désigne la fonction modale qui retourne l'estimateur le plus fréquement rencontré dans l'ensemble des estimateurs. On peut récrire cette fonction comme ceci :

$$
mode(X) = max(count((x_{i})_{i \in \mathbb{N}}))
$$

Exemple : $mode({1,1,2,3,2,2,2,4,1}) = 2$

#### 1.2 Random forest en régression

$$
\hat{y} = \frac{1}{N}\sum_{i = 1}^{N}\hat{y_{i}}
$$

## 2. Exemple en Python

### 2.1 Import des librairies

```python
import pandas as pd

from sklearn.ensemble import RandomForestClassifier  # pour classification
from sklearn.ensemble import RandomForestRegressor   # pour régression
from sklearn.datasets import load_iris               # exemple de dataset
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score           # Métrique de classification
from sklearn.metrics import mean_squared_error       # Métrique de régression
```

### 2.2 Import des données
On reprend le jeu de données iris.

```python
data = load_iris()

df = pd.DataFrame(data.data, columns = data.feature_names)
df["target"] = data.target

X = data.data # Features
y = data.target # Classes

df.head()
```

### 2.3 Séparation des données de test et d'entraînement

```python
# Séparer données en train / test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

### 2.4 Classifier
Pour des problèmes de classification :
#### 2.4.1 Entraînement du modèle

```python
ClassifierModel = RandomForestClassifier(n_estimators=100, random_state=42) # Initialisation du modèle
ClassifierModel.fit(X_train, y_train) # Entraînement du modèle
```

#### 2.4.2 Prédictions

```python
y_pred_class = ClassifierModel.predict(X_test)
```

Pour faire de nouvelles prédictions :

#### 2.4.3 Evaluation des performances du modèle

```python
accuracy_score(y_test, y_pred_class)
```

### 2.5 Entraînement du modèle Regressor
Pour des problèmes de régression :
#### 2.5.1 Entraînement du modèle

```python
RegressorModel = RandomForestRegressor(n_estimators=100, random_state=42)
RegressorModel.fit(X_train, y_train)
```

#### 2.5.2 Prédictions

```python
y_pred_reg = RegressorModel.predict(X_test)
```

```python
Pour faire de nouvelles prédictions :
```

```python
X_new = [[5.1, 3.5, 1.4, 0.2]]  # même structure que les colonnes de X
RegressorModel.predict(X_new)  # ou reg.predict(X_new)
```

```python
X_new = [[5.1, 4.5, 1.4, 0.3]]  # même structure que les colonnes de X
RegressorModel.predict(X_new)  # ou reg.predict(X_new)
```

### 2.7 Evaluation des performances du modèle

```python
mean_squared_error(y_test, y_pred_reg)
```
