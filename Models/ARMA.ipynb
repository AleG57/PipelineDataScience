# Modèle autorégressif (ARMA)
---
## 1. Formulation mathématique

Un modèle ARMA est une combinaison linéaire du modèle AR et du modèle MA. Il est utilisé sur des séries **stationnaire ou non**, i.e. $\mathbb{E}(X_{t})$ pas forcément constant. Ce modèle base ses estimations sur les erreurs et la relation linéaire des valeurs du passé.
