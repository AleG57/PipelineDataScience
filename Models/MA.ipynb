# Modèle autorégressif (AR)
---
## 1. Formulation mathématique

Un modèle MA ou Moving Average est un modèle de régression pour les séries temporelles (variables qui évoluent au cours du temps) **stationnaire uniquement**, i.e. $\mathbb{E}(X_{t}) = cste$. Ce modèle base ses estimations sur les erreurs du passé. On estime $X_{t}$ (futur) en fonction du bruit passé.
