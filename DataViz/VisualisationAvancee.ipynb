On peut directement jouer avec les données brutes grâce aux charts interractifs de la librairie `pywedge`. C'est une librairie puissante qui permet de tracer des graphiques de différents types en fonction de variables du dataset à spécifier.

Attention cependant `pywedge` est un peu capricieux : il faut nettoyer les données avant d'afficher un quelconque graphe et s'assurer que toutes les librairies additionnelles soient installées.

```python
# Import des librairies
import seaborn as sns
import pywedge as pw

from sklearn.preprocessing import LabelEncoder
```

```python
df = sns.load_dataset("titanic") # Import du dataset

# Pré-traitement des données à effectuer pour utiliser pywedge à son plein potentiel
# Retire les lignes qui ont des valeurs manquantes
df.dropna(inplace=True)

# Encode les covariables catégoricielles (i.e. convertit les catégories en valeurs numériques)
for col in df.columns:
    if df[col].dtype == 'object' or str(df[col].dtype) == 'category':
        le = LabelEncoder()
        df[col] = le.fit_transform(df[col])
        
# Pywedge
charts = pw.Pywedge_Charts(df, c=None, y = 'pclass')
plots = charts.make_charts()
```
